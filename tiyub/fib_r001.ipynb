# Data intake and processing

```python
# Define the instruments to download data of a stock/ETF .
tickers = 'AAPL'

# Fetch SPY Data from 1/1/1997 until 12/31/2021.
start_date = datetime.date(2017,1,1)
end_date = datetime.date(2021,12,31)
interval = '1d'
```

```python
# Use pandas_reader.data.DataReader to load the desired data. 
yf.Tickers(tickers)
panel_data = yf.download(
    tickers,
    start = start_date,
    end = end_date,
    interval = interval
)

# Checkout the data type
type(panel_data)
```

```python
# Review data
panel_data # ?: date column not visible in VScode 
```

```python
# remove Adj Close column
ohlc_data = panel_data.drop(columns='Adj Close')

ohlc_data
```

```python
# info() used to determine dtype, NaN count
ohlc_data.info()
```

# FINTA: Fibonacci application

```python
# apply FINTA Fibonacci to df, display head, tail, shape
fib_finta_df=TA.PIVOT_FIB(ohlc_data)

fib_finta_df
```

```python
# clean data
fib_df=fib_finta_df.dropna()

fib_df
```
